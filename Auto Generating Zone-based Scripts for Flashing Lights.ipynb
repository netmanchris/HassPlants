# Auto Generating Light Flashing Scripts

```python
import pandas as pd
from jinja2 import Template
import os
```

```python
manual_sensors = pd.read_csv('data/room_sensors.csv', index_col=0)
```

```python
manual_sensors
```

```python
rooms_list = list(manual_sensors.room.unique())
rooms_list
```

```python
with open('templates/scripts.j2') as f:
    script_template = f.read()
j2_script_template = Template(script_template)
```

```python
# Includes the manual plants
# Includes the LilyGo plants

for room in rooms_list:
    if type(room) is str:  
        room = [room]
        room_zone_df = manual_sensors[manual_sensors.room.isin(room)]
        #print (room_zone_df)
        if len(room_zone_df) > 0:
            #print (room)
            #room_df = room_df[room_df.sensor_model.eq('Miflora')]
            room_dict = room_zone_df.to_dict('records')
            #print (room_dict)
            #print (len(room_dict))

            # OUTPUT A SINGLE FILE FOR EACH ROOM WITH ALL THE MiFLORA PLANTS INTEGRATION YAML WITH THE PROPER MIFLORA SENSOR NAMES
            with open('output/scripts/'+room_dict[0]['room']+'.txt', 'w') as f: 
                f.write(j2_script_template.render( rooms = room_dict))
```
